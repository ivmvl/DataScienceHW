**Исследование каталога украшений бренда Cartier**

В данном исследовании я планирую рассмотреть и проанализировать каталог украшений Cartier: ассортимент, стоимость, состав. Эта информация позволит определить, к какому ценовому сегменту относится бренд, что  является обязательным этапом при разработке маркетинговой стратегии продвижения.


Для начала выгрузим весь каталог:

```python
import pandas as pd
import numpy as np
data = pd.read_csv('cartier_catalog.csv')
data
#Выгружаем исходные данные датасета
```

**Кольца**

Изучим стоимость колец Cartier и определим самую дорогую и самую дешевую модель.

```python
data_tall = data[data['categorie'] == 'rings']
data_tall[['title','tags','price']]
#Сортируем данные по категории "кольца"
```

**Что видим?**

Рассмотрим кольца по критериям состава и цены для того, чтобы проследить зависимость цены от состава. Итак, и самое дешевое, и самое дорогое кольцо выполнены из желтого золота. Судя по данным из нашей таблицы, самая дорогая модель кольца - **Clash de Cartier**, а самая бюджетная - **Love**.

Далее предлагается определить, какое украшение из **всего ассортимента** Cartier является самым дорогим, а какое - самым дешевым.

```python
data.sort_values('price', ascending=True)
#Сортируем данные по возрастанию цены
```

**Что видим?**

Итак, наименьшей ценой (500 у.е.) обладает моно-серьга бренда, а самое дорогое изделие - колье из белого золота с бриллиантами (370 000 у.е.).

**Браслеты**

Теперь рассмотрим ассортимент браслетов бренда Cartier.

```python
group = data.groupby('categorie')
group.get_group('bracelets')
#Группируем данные по браслетам для более детального рассмотрения этой категории.
```

**Что видим?**

К покупке доступны изделия из различных металлов: желтое золото, розовое золото, белое золото. Декорированы браслеты могут быть разными камнями, например, бриллиантами, амазонитами, малахитом и жемчугом.

Теперь определим, сколько в среднем стоит украшение Cartier, какая цена в нашем каталоге встречается чаще всего и посмотрим медиану по таблице.

```python
N = data.price.shape[0]
sum = 0
for i in range(N):
    sum += data.price[i]
mean = float(sum/N)
print("Среднее значение :",mean)
Data1 = data.sort_values('price', ascending=True)
print("Медиана :",Data1.price[N/2])

price_counts = {}
for p in Data1.price:
    if p not in price_counts:
        price_counts[p] = 1
    else:
        price_counts[p] += 1

maxp = 0
mode_price = None
for k, v in price_counts.items():
    if maxp < v:
        maxp = v
        mode_price = k
print("Мода :",mode_price,"Кол-во :", maxp)

print("Меры разброса (также максимальные и минимальные значения) : ",max(Data1.price),"; ",min(Data1.price))

std = float(np.std(Data1.price))
print("Стандартное отклонение : ", round(std,2))

Three = 3
o = ""
zmark = []
for i in Data1.price:
    z = (i - mean) / std
    zmark.append((i - mean) / std)
    if z > Three:
        o += str(i) + ", "
if o == "":
    print("Нет выбросов")
else:
    print(o)
for i in range(len(zmark)):
    print("Z-оценка [",i,"] элемента : ", round(zmark[i],2))
```

Вывели **среднюю цену** (27 057), **медиану** (11 800), **моду** (3 500), а также **меру разброса** - максимальное и минимальное значения в выборке (370 000 и 500).

Ориентируясь на **среднюю цену**, можно отнести украшения Cartier к **премиальному** ценовому сегменту, однако наибольшая цена из прайс-листа позволяет отнести бренд к ценовому сегменту **"pure luxury"**

Кроме того, в нашей базе данных наблюдается большое количество **выбросов**, что позволяет сделать вывод о том, что не все этапы из исследования могут быть использованы для определения ценового сегмента бренда Cartier.
